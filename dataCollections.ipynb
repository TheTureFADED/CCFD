```python
# Improved heatmap with upper-triangle mask and clearer formatting
correlation_matrix = df.corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(18, 16), dpi=120)
sns.heatmap(
    correlation_matrix,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.7, "label": "Correlation"},
    vmin=-1,
    vmax=1,
    annot_kws={"size": 8}
)

plt.title("Correlation Heatmap (Upper Triangle)", fontsize=20, fontweight="bold")
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Visualize class imbalance with percentages
df['Class'] = df['Class'].astype(int)  # Ensure 'Class' is integer type
class_counts = df['Class'].value_counts()
class_percentages = (class_counts / len(df)) * 100

fig, ax = plt.subplots(figsize=(10, 6), dpi=120)
bars = ax.bar(class_counts.index, class_counts.values, color=['#2ecc71', '#e74c3c'], alpha=0.8, edgecolor='black', linewidth=1.5)

# Add value labels on bars
for i, (bar, count, pct) in enumerate(zip(bars, class_counts.values, class_percentages.values)):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(count)}\n({pct:.2f}%)',
            ha='center', va='bottom', fontsize=12, fontweight='bold')

ax.set_xlabel('Class (0=Legitimate, 1=Fraud)', fontsize=12, fontweight='bold')
ax.set_ylabel('Count', fontsize=12, fontweight='bold')
ax.set_title('Class Imbalance Analysis - Credit Card Fraud Detection', fontsize=14, fontweight='bold')
ax.set_xticks([0, 1])
ax.set_xticklabels(['Legitimate', 'Fraud'])
ax.grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```
