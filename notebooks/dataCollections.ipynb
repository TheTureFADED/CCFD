## Amount Distribution by Class

Below we plot the given amounts in distribution plots. This feature is significant as it is the only untransformed feature besides time. We expect that a fraud will typically use a higher amount than regular transactions.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Linear scale
small_tx = df[df['Amount'] <= 500]
sns.histplot(data=small_tx, x='Amount', hue='Class', bins=40, ax=axes[0],
             palette={0: '#2ecc71', 1: '#e74c3c'}, stat='density',
             common_norm=False, element='step', alpha=0.5)
axes[0].set_title('Transaction Amount ≤ $500 by Class')
axes[0].set_xlabel('Amount ($)')

# Log scale
df_plot = df.copy()
df_plot['LogAmount'] = np.log1p(df_plot['Amount'])  # log1p handles zeros gracefully

sns.histplot(data=df_plot, x='LogAmount', hue='Class', bins=40, ax=axes[1],
             palette={0: '#2ecc71', 1: '#e74c3c'}, stat='density',
             common_norm=False, element='step', alpha=0.5)
axes[1].set_title('Transaction Amount by Class (log-transformed)')
axes[1].set_xlabel('log(1 + Amount)')

plt.tight_layout()
plt.show()

print(df.groupby('Class')['Amount'].agg(['mean', 'median', 'max', 'count']).round(2))
```

From the above plots we can notice a large number of fraudulent activity takes place with very low dollar amount transactions. This is due to the idea of card-testing, where a fraudster first verifies the usage of a stolen card on a small purchase. Furthermore, we note that there is much overlap throughout the amount ranges, this implies we will need many strong features to create a coherent prediction.

# All V Features
Below all box plots of V features are shown for additional detail.

```python
features = df.columns.drop('Class')

n_features = len(features)
n_cols = 4
n_rows = (n_features + n_cols - 1) // n_cols

plt.figure(figsize=(18, n_rows * 4))

for i, column in enumerate(features):
    plt.subplot(n_rows, n_cols, i + 1)
    sns.boxplot(
        x='Class',
        y=column,
        hue='Class',
        data=df,
        palette={0: 'lightgreen', 1: 'lightcoral'},
        legend=False,
        showfliers=False
    )
    plt.title(f'{column} by Class', fontsize=11)
    plt.xlabel('')

plt.suptitle('Boxplots of All Features by Class', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()
```
